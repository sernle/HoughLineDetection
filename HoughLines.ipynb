```python
# Follow Udacity Nano AI Course
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/mobile-phone-dream.jpg')

# make copy
image_copy = np.copy(image)

# Change color to RGB 
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# Convert image to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# Define our parameters for Canny
low_threshold = 50
high_threshold = 100

edges = cv2.Canny(gray, low_threshold, high_threshold)

plt.imshow(edges, cmap='gray')


```

```python
# Find lines using a Hough transform
rho = 1  # resolution 1 pixel
theta = np.pi/180  # 1 degree
threshold = 60  # minimum threshold to find the line. Minimum Hough intersection
min_line_length = 100 # minimum line links
max_line_gap = 5 # max line gap


lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

line_image = np.copy(image_copy)

# Iterate over output "lines" and draw lines on image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 5)
plt.imshow(line_image)
```

```python
# Read in an image and convert to RGB
image = cv2.imread('images/road3.jpeg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Convert image to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# Peform edge detection with Canny
low_threshold = 90
high_threshold = 180
edges = cv2.Canny(gray, low_threshold, high_threshold)
# ---------------------------------------------------------- #


##  Define the Hough transform parameters 
##  so that lane lines are the only ones detected

rho = 1
theta = np.pi/180
threshold = 50
min_line_length = 65
max_line_gap = 4


line_image = np.copy(image_copy) #creating an image copy to draw lines on

# Run Hough on the edge-detected image
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                        min_line_length, max_line_gap)


# Iterate over the output "lines" and draw lines on the image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),3)

# ---------------------------------------------------------- #
plt.imshow(line_image)


```
